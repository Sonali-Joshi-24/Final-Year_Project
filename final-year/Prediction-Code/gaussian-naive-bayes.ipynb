```python
# Import necessary modules
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report

train_data = pd.read_csv('traindata.csv')
test_data = pd.read_csv('testdata.csv')
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
X_train = train_data.drop(['law_cat_cd', 'ofns_desc'],axis=1).values
y_train = train_data.get(['law_cat_cd', 'ofns_desc']).values
X_test = test_data.drop(['law_cat_cd', 'ofns_desc'],axis=1).values
y_test = test_data.get(['law_cat_cd', 'ofns_desc']).values
```

```python
def makePredictions(index):
    gnb = GaussianNB()
    gnb.fit(X_train, y_train[:, index])
    
    # making predictions on the testing set
    y_pred = gnb.predict(X_test)
    
    print(classification_report(y_test[:, index],y_pred))
```

```python
makePredictions(0)
```
